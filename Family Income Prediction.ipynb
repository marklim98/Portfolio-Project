```python
df2 = df1.copy()
df2.head()
```

```python
def onehot_encode(df, column, prefix):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
#OneHotEncode categorical variables
nominal_features = [
    ("Region", "reg"),
    ("Main Source of Income", "src_income"),
    ("Household Head Marital Status", "status"),
    ("Household Head Highest Grade Completed", "education"),
    ("Type of Household", "household"),
    ("Type of Building/House", "bldg"),
]
```

```python
for column, prefix in nominal_features:
    df2 = onehot_encode(df2, column=column, prefix=prefix)
```

```python
df2
```

```python
y = df2["Total Household Income"]
X = df2.drop("Total Household Income", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, shuffle=True, random_state=1)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = pd.DataFrame(scaler.transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "L2-Regularized Linear Regression": Ridge(),
    "L1-Regularized Linear Regression": Lasso(),
    "Huber (Robust) Regression": HuberRegressor(),
    "Linear Kernel Support Vector Machine": LinearSVR(),
    "Decision Tree": DecisionTreeRegressor()
}
for name, model in models.items():
    model.fit(X_train, y_train)
    print(name + " Trained")
```

```python
#RESULTS
print("Model R^2 Scores (Test Set):")
for name, model in models.items():
    print(name + ": {:.4}".format(model.score(X_test, y_test)))
```

```python
#Optimization
l2_reg_strength = 1.0

l2_model = Ridge(alpha=l2_reg_strength)
l2_model.fit(X_train, y_train)

print("Ridge Regression Test R^2 (alpha={}): {:.5f}".format(l2_reg_strength, l2_model.score(X_test, y_test)))
```
